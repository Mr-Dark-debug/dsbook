```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dia=pd.read_csv('/content/sample_data/diabetes.csv')
df = pd.DataFrame(dia)
df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['BMI'], df['BloodPressure'], c=df['Age'], cmap='viridis')
plt.colorbar(label='Age')
plt.xlabel('BMI')
plt.ylabel('Blood Pressure')
_ = plt.title('Blood Pressure vs BMI by Age Group')
```

```python
#scatter plot
x = df['Age']
y = df['BloodPressure']
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', alpha=0.4)
plt.xlabel("Age", fontsize=15)
plt.ylabel("Blood Pressure", fontsize=15)
plt.title("Age vs. Blood Pressure", fontsize=15)
plt.grid(True)
plt.show()
```

```python
#Pie chart
diabetes_counts = df['Outcome'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(6, 6))
plt.pie(diabetes_counts, labels=diabetes_counts.index, autopct="%1.1f%%", startangle=140, colors=['lightblue', 'coral'])
plt.title("Distribution of Diabetes Cases", fontsize=15)
```

```python
dia.plot()
plt.xlabel("age",fontsize=15,color="red")
plt.ylabel("diabetes",fontsize=15,color="red")
plt.title("age vs diabetes",fontsize=15,color="red")
plt.legend(fontsize=8,loc="best",)
plt.grid()
plt.show()
```
